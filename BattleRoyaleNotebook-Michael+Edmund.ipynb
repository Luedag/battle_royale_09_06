# Final Model Code

```python
import pandas as pd # data science essentials
from sklearn.model_selection import train_test_split # train/test split
import numpy as np #tools for log transformation
from sklearn.linear_model import LinearRegression #linear modeling package
from catboost import CatBoostRegressor #best gradient regression package
from contextlib import contextmanager #first package for output suppression
import sys, os #second package for output suppression
from sklearn.model_selection import cross_val_score
from skopt.space import Real, Integer
from skopt.utils import use_named_args
file = 'battle_royale_train.csv'
df = pd.read_csv(file)
# preparing response variable data
df_target = df.loc[:, 'PRICE']
#chef_target = np.log(chef_target) #log transformation
#preparing explanatory variable data
df_data   = df.drop(["PRICE"],axis=1)
cat = CatBoostRegressor(depth=7)
cat.fit(df_data,df_target)
df_data['CatBoost'] = cat.predict(df_data)
lr = LinearRegression()
lr.fit(df_data,df_target)
lr.predict(df_data)
lr.score(df_data,df_target).round(4)
#use the code below to score the holdout data
#lr.score(holdout_x,holdout_y).round(4)
```

# Feature Analysis

```python
coeff_df = pd.DataFrame(lr.coef_, df_data.columns, columns=['Coefficient'])  
coeff_df.sort_values(by = 'Coefficient', ascending = False)
```

Our OLS regression model has most of it's feature weighting on the CatBoost predictions and nitrous oxide concentration per town. This mainly indicates that for some reason the CatBoost model under-captures the importance of NOX. Additionally, we must dig further into the feature weights of the catboost model to understand the weighting of our linear regression.

```python
coeff_df = pd.DataFrame(cat.get_feature_importance(),df_data.drop(["CatBoost"],axis=1).columns, columns=['Coefficient'])
coeff_df.sort_values(by = 'Coefficient', ascending = False)
```

Based on the feature importance dataframe, which represents how strongly explanatory variables impact our target variable, we can see that RM and LSTAT are by far the most impactful on Boston housing prices.
